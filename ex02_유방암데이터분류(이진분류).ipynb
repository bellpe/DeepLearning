```python
from sklearn.datasets import load_breast_cancer
```

```python
breast_data = load_breast_cancer()
```

```python
breast_data.keys() #데이터의 키값
```

```python
print(breast_data['DESCR'])
```

```python
# 문제와 정답
X = breast_data['data'] #문제데이터
y = breast_data['target']  # 정답
```

```python
X.shape, y.shape
```

```python
# 훈련, 평가 나누기!
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1230)
```

```python
X_train.shape, y_train.shape
#데이터개수, 특성의수
```

### 모델링
1. 모델 구조 설계
2. 모델 학습/평가 방법 설정
3. 모델 학습 및 학습결과 시각화
4. 모델 평가

```python
from tensorflow.keras import Sequential  #모델을 쌓을 뼈대
from tensorflow.keras.layers import InputLayer, Dense  # InputLayer-데이터가 들어올 층,Dense - 쌓을모델
```

```python
# 1. 모델 구조 설계
model = Sequential()  #뼈대
model.add(InputLayer(input_shape=(30, ))) # 입력층       (들어올 데이터의 크기, 학습시킬데이터 특성? 30개
model.add(Dense(units=32,activation='sigmoid' )) #    값을만낫을때 함수
model.add(Dense(units=16,activation='sigmoid' ))
model.add(Dense(units=1,activation='sigmoid' )) # 출력층
```

```python
# 2. 모델 학습/평가 방법 설정
model.compile(loss='binary_crossentropy',
              optimizer='Adam',
              metrics=['accuracy']) 
```

```python
# 3. 학습 및 학습결과 시각화
breast_history = model.fit(X_train, y_train,
                           validation_split=0.2, #학습시키는데 20퍼센트로 검증까지하겠음
                           epochs=500)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(15,5))
plt.plot(range(1,501),
         breast_history.history['loss'], label='loss')  #훈련 데이터 그래프
plt.plot(range(1,501),
         breast_history.history['val_loss'], label='val_loss')  # 검증 데이터 그래프
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.plot(range(1,501),
         breast_history.history['accuracy'], label='accuracy')  #훈련 데이터 그래프
plt.plot(range(1,501),
         breast_history.history['val_accuracy'], label='val_accuracy')  # 검증 데이터 그래프
plt.legend()
plt.show()
```

```python
# 시퀀스 길이 맞추기
X_train_len = [len(doc) for doc in X_train]
# 8000개 뉴그 기사 길이를 전부 리스트에 넣어줌!
```

```python
print('최소', min(X_train_len))
print('최대', max(X_train_len))
print('평균', np.mean(X_train_len))
print('중앙값', np.median(X_train_len))
# 너무 적거나 많지 않은 110~120 정도의 단어를 RNN이 입력받도록 하자!
```

```python
plt.hist(X_train_len, bins=20)
plt.show()
```

```python
from tensorflow.keras.preprocessing import sequence
```

```python
# 데이터 전처리
# 각 기사의 단어 수가 제각각 다르기 때문에 단어의 숫자를 맞춰준다!
X_train_pad = sequence.pad_sequences(X_train, maxlen=120)
X_test_pad = sequence.pad_sequences(X_test, maxlen=120)
```

```python
X_train_pad.shape, X_test_pad.shape
```

```python
X_train_pad[0]
```

```python
X_train_pad_reshape = X_train_pad.reshape(8982, 120, 1)
X_test_pad_reshape = X_test_pad.reshape(2246, 120, 1)
```

```python
X_train_pad_reshape.shape, X_test_pad_reshape.shape
```

###  모델 설정 및 학습

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import InputLayer, Dense, SimpleRNN
```

```python
model = Sequential()
model.add(InputLayer(input_shape=(120, 1)))
model.add(SimpleRNN(128))
model.add(Dense(32, activation='relu'))
model.add(Dense(46, activation='softmax'))
```

```python
model.compile(loss='sparse_categorical_crossentropy',
             optimizer='adam',
             metrics=['accuracy'])
```

```python
history = model.fit(X_train_pad_reshape, y_train, validation_split=0.2,
                   epochs=20)
```

```python
plt.figure(figsize=(15,5))
plt.plot(history.history['accuracy'], label='accuracy')
plt.plot(history.history['val_accuracy'], label='val_accuracy')
plt.show()
```

###  LSTM 모델링

- 보완하기 위해 기억의 중요성을 관리하는 연산이 추가된 LSTM 개발
- 중요한 기억은 오래 보존, 불필요한 기억은 삭제되도록 관리!

### 워드 임베딩
- 현재 로이터 뉴스는 단어를 수치화 할 때 빈도수 기준으로 수치화 해둔 상태!
- 단어의 본질적인 의미를 이해하기 위해서는 단순히 단어를 구분하는 수치화가 아니라. 단어의 의미에 따라 세부적으로 만들어진 실수형태의 표현법이 필요하다!
- 단어 사이의 관계를 풍부하게 밀집된 표현으로 만들 수 있는 워드 임베딩을 사용해보자!


```python
from tensorflow.keras.layers import Embedding
```

```python
model3 = Sequential()
model3.add(Embedding(1000,50)) #사용하는 단어의 수, 각 단어를 표현할 숫자의 수
model3.add(LSTM(128))
model3.add(Dense(32,activation='relu'))
model3.add(Dense(46,activation='softmax'))
```

```python
model3.compile(loss='sparse_categorical_crossentropy',
             optimizer='adam',
             metrics=['accuracy'])
```

```python
history3 = model3 .fit(X_train_pad_reshape, y_train,
                      validation_split=0.2, epochs=20)
```

```python
plt.figure(figsize=(15,5))
plt.plot(history3.history['accuracy'], label='accuracy')
plt.plot(history3.history['val_accuracy'], label='val_accuracy')
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.plot(history3.history['loss'], label='loss')
plt.plot(history3.history['val_loss'], label='val_loss')
plt.show()
```

###  모델 수정해보기
- num_words 숫자 -> Embedding 숫자 수정
- 시퀀스 길이 설정
- LSTM 뉴런 갯수 / LSTM층의 깊이

```python
from tensorflow.keras.layers import Embedding
```

```python
model4 = Sequential()
model4.add(Embedding(1000, 50  )) #사용하는 단어의 수, 각 단어를 표현할 숫자의 수
model4.add(LSTM(128, return_sequences=True))  # 다수 입력 다수 출력으로 바꿔줘야 함
model4.add(LSTM(128, return_sequences=True)) 
model4.add(LSTM(128))   # tanh 활성화함수
model4.add(Dense(32, activation='relu'))
model4.add(Dense(46,activation='softmax'))
```

```python
model4.compile(loss='sparse_categorical_crossentropy',
             optimizer='adam',
             metrics=['accuracy'])
```

```python
history4 = model4 .fit(X_train_pad_reshape, y_train,
                      validation_split=0.2, epochs=20)
```

```python
plt.figure(figsize=(15,5))
plt.plot(history4.history['accuracy'], label='accuracy')
plt.plot(history4.history['val_accuracy'], label='val_accuracy')
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.plot(history4.history['loss'], label='loss')
plt.plot(history4.history['val_loss'], label='val_loss')
plt.show()
```

### 이진(Binary) 이미지 만들기

```python
img = cv2.imread('images/jimin.jpg', cv2.IMREAD_GRAYSCALE) 

# 이진이미지 변환
ret, thr1 = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)

# 127아래면 0 검정처리, 127위면 255 하얀색처리해줌
# threshold 임계값, 배열이미지 값 두가지 반환

plt.xticks([])
plt.yticks([])
plt.imshow(thr1, cmap='gray') 
plt.show()              
```

- cv2.threshold(img, threshold_value, value, flag)
   - img : gray이미지
   - threshold_valuie : 임계값 (0, 1로 나누기 위한 값)
   - value : 임계값보다 클 때 적용되는 값
   - flag : 적용방법

### 비디오 프레임 출력하기

```python
cap = cv2.VideoCapture('images/video.mp4') #불러옴 , 읽어온 비디오를 cap변수에 담음

#영상은 이미지의 연속
#윈도우상에서 열기

while True:  #반복문 통해서 조건이 될때 계속 출력시킴
    ret, frame = cap.read()  # ret, 이미지 불러오는 성공여부
    # ret - T/F (read의 성공여부)
    
    if not ret:  # 이미지 불러오지 못하면 T! → 조건문 출력
        print('비디오 읽기 오류')
        cap.release()
        cv2.destroyAllWindows()
        break
        
    cv2.waitKey(1) # 1이면 한 사진을 0.001ms로 보여주기
    cv2.imshow('video', frame)
```

### 비디오 녹화하기

```python
cap = cv2.VideoCapture('images/video.mp4') # 동영상 
#cap = cv2.VideoCapture('images/video.mp4')

fps = 30.0 # fram per secnod
width = int(cap.get(3)) # 비디오 가로
height = int(cap.get(3)) # 비디오 세로

fcc = cv2.VideoWriter_fourcc(*'MPEG')  # 코덱

out= cv2.VideoWriter('video01.mp4', fcc, fps,(width, height))
# 저장시킬 이름, 코덱 정보, 프레임, 영상 크기

record = False  # record에 false 넣어줌

while True:
    ret, frame = cap.read() 
    # ret - T/F (read의 성공여부)
    
    if not ret:   # 이미지 불러오지 못하면 T -> 조건문 실행!
        print('비디오 읽기 오류')
        cap.release()
        cv2.destroyAllWindows()
        break
        
    cv2.imshow('vieo',frame)
    
    if record : 
        out.write(frame)
    
    k = cv2.waitKey(30) # 1이면 한 사진을 0.001ms로 보여주기
    
    if k== 49:  # 키보드 1 누르면 비디오 닫기
        cap.release()
        cv2.destroyAllWindows()
        break
        
    if k == 50:   # 키보드 2 누르면 녹화 시작
        print('녹화시작')
        record = True
        
        
    if k == 51:  #  키보드 3 누르면 녹화 종료
        print('녹화종료')
        record = False
        out.release()
```

### 비디오의 한 프레임 이미지 파일로 저장 (캡처)

```python
cap = cv2.VideoCapture('images/video.mp4') # 동영상 

cnt=0

while True:
    fet, frame = cap.read()
    
    k = cv2.waitKey(30)
    
    if k == 50:  # 키보드 2를 누르면 캡쳐
        cnt += 1
        print('test{}.png'.format(cnt))
        cv2.imwrite('test{}.png'.format(cnt), frame)
        
    cv2.imshow('video', frame) 
    
    if k == 49:
        cap.release()    
        cv2.destroyAllWindows()
        break
```
